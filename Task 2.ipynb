## Importing titanic dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("titanic.csv")
df.head()
```

## Understanding the dataset

```python
df.info()
```

```python
df.describe()
```

## Checking missing value

```python
df.isnull().sum()
```

```python
df.drop('Cabin',axis=1, inplace = True)
```

```python
df['Age'].fillna(df['Age'].median(),inplace = True)
df['Embarked'].fillna(df['Embarked'].mode(),inplace =True)
```

```python
df.head()
```

```python
df.value_counts()
```

```python
# Convert 'Sex' and 'Embarked' to categorical data type
df['Sex'] = df['Sex'].astype('category')
df['Embarked'] = df['Embarked'].astype('category')

# Convert 'Survived' to categorical
df['Survived'] = df['Survived'].astype('category')
```

```python
df
```

## Performing EDA

```python
#Distribution of passengers by Pclass
sns.countplot(x='Pclass', data=df, hue='Survived')
plt.title('Passenger Pclass Distribution by Survival')
plt.show()
```

```python
# Create a count plot for 'Survived' with hue='Sex'
plt.figure(figsize=(8, 6))
sns.countplot(x='Survived', hue='Sex', data=df)
plt.title('Survival Count by Sex')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.legend(title='Sex', loc='upper right')
plt.show()
```

```python
#creating kde plot for 'Survived' with hue='Age'
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='Age', hue='Survived', fill=True, alpha=0.5)
plt.title('KDE Plot of Age by Survival')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend(title='Survived', labels=['Not Survived', 'Survived'])
plt.show()
```

```python
# Distribution of passengers by Embarked
# this is useful for understanding if there's any correlation between the port of embarkation and survival.
plt.figure(figsize=(8, 6))
sns.countplot(x='Embarked', data=df, hue='Survived')
plt.title('Passenger Embarked Distribution by Survival')
plt.xlabel('Embarked')
plt.ylabel('Count')
plt.show()
```

```python
# Distribution of passengers by SibSp
#shows the distribution of passengers based on the number of siblings/spouses (SibSp) onboard. 
#It helps in understanding how having siblings or spouses onboard might have influenced survival.
plt.figure(figsize=(8, 6))
sns.countplot(x='SibSp', data=df, hue='Survived')
plt.title('Passenger SibSp Distribution by Survival')
plt.xlabel('SibSp')
plt.ylabel('Count')
plt.show()
```

## Feature engeenering

```python
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
```

## Statistical Analysis

```python
# Correlation matrix
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
# Pairplot
sns.pairplot(df[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']])
plt.title('Pairplot of Selected Features')
plt.show()
```

# Summary of Findings

## Data Cleaning Observations:
- **Missing Values**: No missing values were observed in the dataset.
- **Outliers**: No explicit examination of outliers was conducted in this analysis.

## Key Insights:
- **Embarked Distribution**: Passengers predominantly embarked from Southampton ('S'), followed by Cherbourg ('C') and Queenstown ('Q').
- **Survival Rates**: Survival rates varied across different passenger classes, sexes, and age groups. Female passengers had higher survival rates compared to males, and passengers in higher passenger classes (Pclass) tended to have better chances of survival.
- **Impact of Age**: There seems to be a varied impact of age on survival. While children and younger adults had relatively higher survival rates, there were also instances of survival among older passengers.
